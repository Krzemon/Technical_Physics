```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt

m = 1; v0 = 0; x0 = 2.8; dt1 = 0.01; dt2 = 0.001; T = 30                # w jednostkach SI
t1 = np.arange(dt1,T + dt1, dt1); t2 = np.arange(dt2,T + dt2, dt2)
x = sp.Symbol('x')
V = -sp.exp(-x**2) - 1.2*sp.exp(-(x-2)**2)
dV_dx = sp.diff(V,x)
Vn = sp.lambdify(x, V)
dV = sp.lambdify(x, dV_dx)
```

```python
def x(x0,v0,m,dt1,t1,alfa):   #jesli jest bez tlumienia, to alfa = 0
    v_t = [v0]
    x_t = [x0]
    for i in range(len(t1)-1):
        dV = dV_dx.subs('x', x_t[i])
        v_t.append(v_t[i] - 1/m * dV * dt1 - alfa * v_t[i] * dt1)
        x_t.append(x_t[i] + v_t[i]*dt1)
    return [x_t, v_t]
```

```python
ddV = lambdify(x, V.diff(x), 'numpy')
dt = 0.01
alfa = 0
T = 100
n = 0
A = np.zeros((20,2,2))
B = np.ones((20,2,1))

def matrix(dt,x0,v0,x,v,a,i):
    v_t = [v0]
    x_t = [x0]
    for j in range(i):
        #dVx = dV.subs('x', x_t[j])
        dVx = ddV(x_t[j])
        v_t.append(v_t[j] - 1/m * dVx * dt - alfa * v_t[j] * dt)
        x_t.append(x_t[j] + v_t[j] * dt)
        
    
    #dVx = dV.subs('x', x_t[i])
    #dV0 = dV.subs('x', x_t[0])
    dVx1 = ddV(x_t[i])
    dV0 = ddV(x_t[0])
    v_t.append(v_t[i] - 1/m * dVx1 * dt - alfa * v_t[i] * dt)
    x_t.append(x_t[i] + v_t[i] * dt)
    
    A[i][0][0] = 1
    A[i][0][1] = -dt/2
    A[i][1][0] = dt/(2*m) * dVx1
    A[i][1][1] = 1 + a*dt/2

    B[i][0][0] = x - x0 - dt/2 * v - dt/2 * v0
    B[i][1][0] = v - v0 -dt/2 * (-1/m * dVx1 - a * v) - dt/2 * (-1/m * dV0 - a * v0)
    
    return np.linalg.solve(A[i],B[i])
```

```python
z = 1
y = 1
print("Warość początkowa x i v:",z,y)
for i in range(4):
    b12 = matrix(dt,x0,v0,z,y,0,i)
    z = z - b12[0]
    y = y - b12[1]
    print("x i v:",z,y, "z różnicą dx i dv",b12[0],b12[1])
```

```python
def daneFun(T,dt,a):
    time = int(T/dt)
    z=0
    y=0
    x0=2.8
    v0=0
    xx = [];yy = [];EE = [];VV = [];EEVV = []
    for i in range(time):
        for j in range(5):
            b12 = matrix(dt,x0,v0,z,y,a,j)
            z = z - b12[0]
            y = y - b12[1]
        x0 = z
        v0 = y
        xx.append(z)
        yy.append(y)
        EE.append(m/2 * y**2)
        Vn = V.subs('x', z)
        VV.append(Vn)
        EEVV.append(m/2 * y**2 + Vn)
    return [xx,yy,EE,VV,EEVV]
```
