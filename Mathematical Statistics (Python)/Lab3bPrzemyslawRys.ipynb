```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math as math

x = 1
data_1 = []
srednia_1 = []
numData = 1000
for i in range(numData):
    data_1 = stats.triang.rvs(c=0.5,  loc=-1,  scale=2, size = x)
    srednia_1.append(np.mean(data_1))
plt.hist(srednia_1,bins=40,range=(-1,1),histtype='stepfilled', alpha=0.3, density=True)
```

```python
#x=1
stats.probplot(srednia_1, plot=plt)
plt.show()
print("Test Kołmogorow-Smirnov:")
s_1,p_1 = stats.kstest((srednia_1-np.mean(srednia_1))/((np.std(srednia_1,ddof=1))), 'norm') 
p_1
```

```python
xP = 5
P1_5 = []
P2_5 = []
for i in range(numData):
    k_5 = stats.poisson.rvs(2,size = xP)
    k_0 = 0
    for j in range(xP):
        if k_5[j]==0:
            k_0 += 1
    P1_5.append(math.exp(-np.mean(k_5)))
    P2_5.append(k_0/numData)
plt.subplot(1, 2, 1)
plt.hist(P1_5)
plt.title('P1_5')
plt.subplot(1, 2, 2)
plt.hist(P2_5)
plt.title('P2_5')
plt.show()

srednia_P1_5 = np.mean(P1_5)
sigma_P1_5 = np.std(np.mean(P1_5),ddof=1)
srednia_P2_5 = np.mean(P2_5)
sigma_P2_5 = np.std(np.mean(P2_5),ddof=1)
print('srednia_P1_5 = {0}'.format(srednia_P1_5))
print('sigma_P1_5 = {0}'.format(sigma_P1_5))
print('srednia_P2_5 = {0}'.format(srednia_P2_5))
print('sigma_P2_5 = {0}'.format(sigma_P2_5))
```
